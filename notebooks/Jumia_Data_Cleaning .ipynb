## Jumia Data Cleaning

```python
#importing library for data cleaning

import pandas as pd
import numpy as np
import re
```

```python
# importing dataset for data cleaning

df = pd.read_csv(r"C:\Users\ikeol\OneDrive\Documents\Web-Scraping\Data\Jumia Phones data.csv")
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
#Removed duplicate values from the dataframe

df = df.drop_duplicates()
df
```

```python
df.info()
```

```python
#Dropped the column "Not_Useful_column" cause of its irrelevance to the dataset

df = df.drop(columns = "Unnamed: 0")
df
```

```python
df.head(50)
```

```python
df["Price"] = df["Price"].str.strip("₦")
df["Old Price"] = df["Old Price"].str.strip("₦")

df.head(30)
```

```python
# making the prices (current and old) in numbers

df["Price"] = df["Price"].str.replace("," , "")
df["Old Price"] = df["Old Price"].str.replace("," , "")

df.head(50)
```

```python
df[df['Price'].str.contains('₦')]
```

```python
# extracting a single price in rows that have two prices

def extract_after_naira(value):
    value = str(value)
    if '₦' in value:
        match = re.search(r'₦\s*([\d,]+)', value)
        return match.group(1).replace(',', '') if match else np.nan
    else:
        return value

df['Price'] = df['Price'].apply(extract_after_naira)
df['Old Price'] = df['Old Price'].apply(extract_after_naira)

df.head(41)
```

```python
df[['Price', 'Old Price']].isnull().sum()
```

```python
#Converting null values to NaN, to make it a recongnised null by pandas

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Old Price'] = pd.to_numeric(df['Old Price'], errors='coerce')
```

```python
df[['Price', 'Old Price']].isnull().sum()
```

```python
#Filling null old prices with current price, meaning there are not discounts on those products

df['Old Price'] = df['Old Price'].fillna(df['Price'])
```

```python
#Converting prices to integer from strings

df['Old Price'] = df['Old Price'].astype(int)
df['Price'] = df['Price'].astype(int)
```

```python
#Checking if error is fixed

df.info()
df.dtypes
```

```python
#Replacing Brand rows with null where value is having reviews instead of brand names

df['Brand'] = df['Brand'].apply(lambda x: np.nan if 'verified' in str(x).lower() else x)

df.head(41)
```

```python
#checking for null brand rows

df[df['Brand'].isna()][['Title', 'Brand']]
```

```python
# Fixing null brand names by identifying brand name from title and removing items that don't have brand names
brand_map = {
    'oppo': 'Oppo',
    'huawei': 'Huawei',
    'vivo': 'Vivo',
    'iphone': 'Apple',
    'xiaomi': 'XIAMi',
}

def fill_brand(row):
    if pd.isna(row['Brand']):
        for keyword, brand in brand_map.items():
            if keyword.lower() in row['Title'].lower():
                return brand
    return row['Brand']

df['Brand'] = df.apply(fill_brand, axis=1)

# dropping any item that has missing brand names even in the title
df.dropna(subset=['Brand'], inplace=True)

df[df['Brand'].isna()][['Title', 'Brand']]
```

```python
df.info()
```

```python
# Clean Rating: Extracting just the rating numbers from the rating column 
df['Rating'] = df['Rating'].astype(str).str.extract(r'([\d.]+)').astype(float)

df.head(10)
```

```python
# Clean Reviews: Extracting just the numbers number of reviews from the review column 

df['Reviews'] = df['Reviews'].astype(str).str.extract(r'([\d,]+)')[0].str.replace(',', '').astype(float)

df.head(10)
```

```python
df.dtypes
```

```python
#Converting reviews column to integer

df['Reviews'] = df['Reviews'].fillna(0).astype(int)
```

```python
#Confirming if reviews have been converted to integer

df.dtypes
```

```python
df.head(10)
```

```python
#Checking for more null values

df.info()
```

```python
df[df['Reviews'].isna()][['Title', 'Reviews']]
```

```python
#Filling empty review rows with 0 because they could be new product listings amongst them

df['Reviews'] = df['Reviews'].fillna(0).astype(int)

df.info()
```
